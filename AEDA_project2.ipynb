```python
def eda():
    import pandas as pd
    import numpy as np
    import seaborn as sb
    import matplotlib.pyplot as plt
    import warnings
    warnings.filterwarnings("ignore")
    %matplotlib inline
    
    def load_data(file_path):
        if file_path.endswith('.csv'):
            return pd.read_csv(file_path)
        elif file_path.endswith('.xlsx'):
            return pd.read_excel(file_path)
        elif file_path.endswith('.sql'):
            # return pd.read_sql(query, connection)
            pass
        else:
            raise ValueError("Unsupported file format")
    
    
    file_name = input('Please enter the path of the file: ')
    dt= load_data(file_name)
        
    print('\n')
    print(dt.info())
    print('\n')
    print("Number of dublicated data: ",dt.duplicated().sum())
    print('\n')
    print('\n')
    if(dt.duplicated().sum()>0):
        dt.drop_duplicates(inplace= True)
        print('data without dublicated\n')
        print(dt)
    print(dt.describe())
    try:
        dt.describe(include=['bool','object'])
    except:
        print("There are not Categoriacl col")
    print("number of missin values in columns :")
    print(dt.isna().sum())
    print("fill missing values with median of dataset: \n")  
    dt.fillna(dt.median(),inplace=True)
    print(dt)
    plt.figure(1, figsize=(15, 5))
    sb.heatmap(dt.corr(), annot=True)
    dt.plot(subplots=True, figsize=(15, 10))
    plt.show()
    for i in dt.columns:
        dt[i].hist()
        plt.xlabel(i)
        plt.show()
eda()
    
```
